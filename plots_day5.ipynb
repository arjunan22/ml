```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')#inbuilt package in seaborn
```

```python
plt.plot(tips['total_bill'],tips['tip'],"bo")
plt.xlabel('My X axis')
plt.ylabel('My Y axis')
plt.title('Plot title here')
```

```python
plt.hist(tips['total_bill'],bins=50)
```

```python
plt.boxplot(tips['total_bill'],showfliers=True)
```

```python
#distplot
#distribution plot-----Histogram One variable
sns.distplot(tips['total_bill'])
```

```python
#jointplot
#at elast 2 variables
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.pairplot(tips,hue="sex")
```
